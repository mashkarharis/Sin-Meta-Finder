```python
converted_song_names = []
from_language, to_language = 'en', 'si'
for name in song_names:
    res = tss.itranslate(name.strip(), from_language, to_language)
    print(res)
    converted_song_names.append(res.strip())
converted_song_names
```

```python
with open('1. song_names_processed.txt', 'w', encoding="utf-8") as f:
    for line in converted_song_names:
        f.write(f"{line}\n")
```
