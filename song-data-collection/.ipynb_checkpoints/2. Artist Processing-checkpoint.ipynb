### Import Libraries

```python
import pandas as pd
import translators as ts
import translators.server as tss
import json
```

```python
songs = pd.read_excel("0. Manually Collected Metahpor Data.xlsx")
songs.head()
```

```python
song_dict = None
with open("1. output.json") as json_file:
    song_dict = json.load(json_file)

links = list(songs['YT'].dropna())
i=0
for link in links:
    song_dict[i]["youtube_link"]=str(link)
    song_dict[i]["name"]=song_dict[i]["name"].strip()
    i+=1
song_dict
```

```python
authors_en = set()

for names in songs['Author'].dropna():
    name_list = names.split("|")
    for name in name_list:
        authors_en.add(name.strip())
authors_en
```

```python
map = {}
from_language, to_language = 'en', 'si'
for name_en in authors_en:
    name_sin = tss.itranslate(name_en.strip(), from_language, to_language).strip()
    map[name_en]=name_sin
map
```
