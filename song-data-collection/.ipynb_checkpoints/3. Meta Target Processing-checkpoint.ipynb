```python
current_song=-1
metas=None
for i in range(len(songs["Target (Upameya)"].dropna())):
    target = resolved[songs["Target (Upameya)"][i]]
    meta_no = songs['Metaphors No'][i]
    if meta_no==1:
        if current_song>=0:
            song_dict[current_song]['metaphors']=metas
        current_song+=1
        metas=[]
    metas.append({
        "meta_id": int(meta_no),
        "targets": [i.strip() for i in target.split("/")]
    })
song_dict[current_song]['metaphors']=metas
song_dict
```

```python
json_object = json.dumps(song_dict)
 
with open("3. output.json", "w",encoding='utf8') as outfile:
    outfile.write(json_object)
```
