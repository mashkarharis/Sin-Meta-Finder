```python
source_en = set()
for t in songs["Source (Upamawa)"].dropna():
    source_en.add(t)
source_en
```

```python
map = {}
from_language, to_language = 'en', 'si'
for name_en in source_en:
    name_sin = tss.itranslate(name_en.strip(), from_language, to_language).strip()
    print(name_sin)
    map[name_en]=name_sin
```
