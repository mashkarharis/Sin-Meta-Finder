```python
current_song=-1
metas=None
for i in range(len(songs["Interpretation"].dropna())):
    inter_list= resolved[songs["Interpretation"][i]]
    meta_no = songs['Metaphors No'][i]
    if meta_no==1:
        current_song+=1
    current_meta = song_dict[current_song]['metaphors'][int(meta_no-1)]
    S = current_meta["source"]
    T = random.choice(current_meta["targets"])
    I = random.choice(inter_list)
    I = I.replace("{T}",T)
    I = I.replace("{S}",S)
    current_meta["interpretation"] = I
song_dict
```
