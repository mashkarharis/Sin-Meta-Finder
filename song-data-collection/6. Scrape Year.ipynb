```python
from youtube_scraping_api import YoutubeAPI
import json
```

```python
song_dict = None
with open("5. output.json",'r') as json_file:
    song_dict = json.load(json_file)
song_dict
```

```python
api = YoutubeAPI()
for song in song_dict:
    yt_id = song["youtube_link"].split("=")[-1]
    v=api.video(yt_id)
    song["view_count"]=v.view_count
    song["published_on"]=v.publish_time
    song["length"]=v.length
    print(v.title)
```

```python
song_dict
```

```python
json_object = json.dumps(song_dict)
 
with open("6. output.json", "w",encoding='utf8') as outfile:
    outfile.write(json_object)
```
